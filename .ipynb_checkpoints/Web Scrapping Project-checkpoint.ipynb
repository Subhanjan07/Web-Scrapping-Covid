```python
import requests
```

```python
html = requests.get(url).content
```

```python
type(html)
```

```python
html
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html,'html.parser')
```

```python
tb = soup.find('tbody')
```

```python
tr = tb.find_all('tr')
```

```python
tr[8]
```

```python
all_data = []
for index in range(8, len(tr)):
    country_data = []
    table_row = tr[index]
    table_data = table_row.find_all('td')
    for item in table_data:
        country_data.append(item.text)
        
    all_data.append(country_data)
```

```python
all_data
```

```python
data = list(map(lambda x: x[1:10] + [x[14]], all_data))
```

```python
data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
y = ['country', 
           'totalcase', 
           'newcase', 
           'totaldeath',
           'newdeath', 
           'totalrecovered', 
           'newrecovered', 
           'activecases', 
           'serious', 
           'population']
```

```python
len(y)
```

```python
df.columns = y
```

```python
df.head(20)
```

```python
df.tail(20)
```

```python
df.to_csv('Covid Data-Jan11.csv', index = False) #comma separated value(CSV)
```

# Inserting csv file into databse

```python
df.head()
```

```python
import sqlite3
```

```python
con = sqlite3.connect('web.db')
```

```python
con
```

```python
# you need cursor to execute sql query
```

```python
cur = con.cursor()
```

```python
cur
```

```python
df.columns
```

```python
table_create_query = f'create table covid_data {tuple(df.columns)}'
```

```python
table_create_query
```

```python
cur.execute(table_create_query)
```

```python
datapoints = df.to_records(index = False).tolist()
```

```python
insert_query = """
insert into covid_data values (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
"""
```

```python
cur.executemany(insert_query, datapoints)
```
